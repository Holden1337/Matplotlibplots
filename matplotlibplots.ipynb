# A collection of plots made with matplotlib, numpy, and scipy

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
%matplotlib inline
from scipy import special
import numpy as np
import seaborn as sns
print("Setup Complete")
```

## First plot: plotting $f(x) = x^2$ 

```python
#making two arrays for function f(x) = x^2
x = np.linspace(0,10,num=1000)
y = [x**2 for x in x]
```

```python
#plotting and making axes and labels and whatnot
plt.figure(1)
plt.figure(figsize=(10,10))
plt.title("Graph of $f(x) = x^2$")
plt.grid(True)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('$f(x)$')

```

## Second plot: plot of $f(x) = sinc(x)$

```python
#make array to plot sinc function
x = np.linspace(-10,10,num=1000)
y = [np.sinc(x) for x in x]

plt.figure(2)
plt.figure(figsize=(10,10))
plt.title("Plot of $f(x) = sinc(x)$")
plt.plot(x,y)
plt.grid(True)
plt.xlabel('x')
plt.ylabel("$f(x)$")
```

## Third plot: plotting multiple curves on the same plot

```python
#first function. y = sin(x)
x1 = np.linspace(-10,10,num=1000)
y1 = [np.sin(x) for x in x1]

#second function. y = cos(x)
x2 = np.linspace(-10,10,num=1000)
y2 = [np.cos(x) for x in x2]

plt.figure(3)
plt.figure(figsize=(15,5)) #change figure size
plt.plot(x1,y1,'k--',label='sin') #plot first function
plt.plot(x2,y2,label='cos') #plot second function 
plt.legend() #create legend
plt.xlabel('x') #make x label
plt.ylabel('$f(x)$') #make y label
plt.title("$\sin(x)$ vs $\cos(x)$") #make title
plt.grid(True) #add grid
```

## Fourth plot: plotting multiple curves on different plots

```python
#first function: bessel function
x = np.linspace(0,10,num=1000)
y = [special.jv(0,x) for x in x]

plt.figure(4)
plt.subplot(2,1,1)
plt.plot(x,y,'k--')

plt.subplot(2,1,2)
plt.plot(x,np.sin(x))

plt.show()
```

## Fifth plot: Lots of bessel functions

```python
#plotting bessel functions of different orders
x = np.linspace(0,10,num=1000)

plt.figure(5)
plt.figure(figsize=(12,6))
plt.grid(True)


for n in range(5):
    y = [special.jv(n,x) for x in x]
    plt.plot(x,y,label="n = "+str(n))
    
plt.legend()
plt.xlabel("x")
plt.ylabel("$f(x)$")
plt.title("Plot of Bessel functions for orders 0-4")
```

## Plot 6: histogram with text added

```python
mu, sigma = 100,15
x = mu + sigma * np.random.randn(100000)

plt.figure(6)
plt.figure(figsize=(7,7))

n,bins,patches = plt.hist(x,50,density=1,facecolor='g',alpha=.75)


plt.xlabel('IQ')
plt.ylabel('Probability')
plt.title("Histogram of IQ")
plt.text(60,.025,r'$\mu=100,\ \sigma=15$')
plt.axis([40,160,0,.03])
plt.grid(True)
plt.show()
```

## Seventh plot: 3d plt

```python


fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

## Eighth figure: Pie chart

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

## Ninth figure: legendre polynomials

```python
x = np.linspace(-1,1,num=1000)

plt.figure(7)
plt.figure(figsize=(12,6))

for n in range(6):
    y = [special.eval_legendre(n,x) for x in x]
    plt.plot(x,y,label="n = "+str(n))
    
plt.legend()
plt.xlabel("x")
plt.ylabel("$f(x)$")
plt.title("Plot of Legendre polynomial for orders 0-5")
plt.grid(True)
```

## tenth figure: Normal distribution


```python
from math import pi, exp, sqrt
def norm(mu,sigma,x):
    return (1/(sigma*sqrt(2*pi)))*exp(-.5*((x-mu)/sigma)**2)
muarr = [0,0,0,-2]
sarr = [.2,1,3,.5]
x = np.linspace(-5,5,num=1000)

plt.figure(8)
plt.figure(figsize=(12,6))

for i in range(4):
    y = []
    for j in range(len(x)):
        y.append(norm(muarr[i],sarr[i],x[j]))
    plt.plot(x,y,label="$\mu$ = " + str(muarr[i]) + ", $\sigma^2 = $ " + str(sarr[i]))
    
plt.legend()
plt.xlabel("$\chi$")
plt.ylabel("$\phi_{\mu,\sigma^2}(\chi)$")
plt.title("A selection of PDFs for the normal distribution for different values of $\mu$ and $\sigma^2$")
plt.grid(True)
```

## Eleventh plot: CDF for normal distribution 

```python
from math import erf
def normcdf(mu,sigma,x):
    return .5*(1+erf((x-mu)/sigma*sqrt(2)))

plt.figure(9)
plt.figure(figsize=(12,6))


for i in range(4):
    y = []
    for j in range(len(x)):
        y.append(normcdf(muarr[i],sarr[i],x[j]))
    plt.plot(x,y,label="$\mu$ = " + str(muarr[i]) + ", $\sigma^2 = $ " + str(sarr[i]))
    
plt.legend()
plt.xlabel("$\chi$")
plt.ylabel("$\Phi_{\mu,\sigma^2}(\chi)$")
plt.title("A selection of CDFs for the normal distribution for different values of $\mu$ and $\sigma^2$")
plt.grid(True)
```

## Twelth plot: Scatter plot with different sized dots

```python
from sklearn.datasets import load_iris #load iris dataset
iris = load_iris()
features = iris.data.T


plt.figure(12)
plt.figure(figsize=(12,6))
plt.scatter(features[0],features[1],alpha = .4,s=100*features[3],c=iris.target,cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.grid(True)
```

## Plot 13: Using Errorbars

```python
plt.figure(13)
plt.figure(figsize=(12,6))
x = np.linspace(0,10,50)
dy = .8
y = np.sin(x)+dy*np.random.randn(50)
plt.errorbar(x,y,yerr=dy,fmt='.k')
plt.xlabel("x")
plt.ylabel('y')
```

## Plot 14: Countour plot for function of two variables

```python
def f(x,y):
    return np.sin(x)**10 + np.cos(10+y*x)*np.cos(x)

x = np.linspace(0,5,50)
y = np.linspace(0,5,50)

X,Y = np.meshgrid(x,y)
Z = f(X,Y)

plt.contour(X,Y,Z,colors='black')
```

## Plot 15: Two-Dimensional histogram

```python
mean = [0,0]
cov = [[1,1],[1,2]]
x,y = np.random.multivariate_normal(mean,cov,100000).T

plt.hist2d(x,y,bins=30,cmap='Blues')
cb = plt.colorbar()
cb.set_label('Counts in bin')
```
